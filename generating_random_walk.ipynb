# Intro

1D and 2D random walks are investigated. The connection of the random walk to central limit theorem is shown

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_random_walk(N, d, low=-0.5, high=0.5):
    X = np.random.uniform(low=low, high=high, size=[N, d])
    X_sum = X.cumsum(axis=0)
    return((X,X_sum))
```

# 1D Random Walk

```python
N = 10
d = 1

np.random.seed(1)
plt.figure(figsize=(12, 3))
for i in [1, 2, 3]:
    plt.subplot(1,3,i)
    _, x = gen_random_walk(N,d)
    plt.plot(x)

plt.suptitle(f'{d}D random walk N={N}')
```

# 2D Random Walk

```python
N = 100
d = 2
np.random.seed(1)
plt.figure(figsize=(12, 3))
for i in [1, 2, 3]:
    plt.subplot(1,3,i)
    _, x = gen_random_walk(N,d)
    plt.scatter(x[:,0], x[:,1], marker='.')

plt.suptitle(f'{d}D random walk N={N}')
```

```python
N = 10000
d = 2
np.random.seed(1)
plt.figure(figsize=(12, 3))
for i in [1, 2, 3]:
    plt.subplot(1,3,i)
    _, x = gen_random_walk(N,d)
    plt.scatter(x[:,0], x[:,1], marker='.')

plt.suptitle(f'{d}D random walk N={N}')
```

```python
W = 10000
d = 2

np.random.seed(1)
plt.figure(figsize=(12, 6))

N = 1
end_points = list()
for i in range(W):
    _,x = gen_random_walk(N,d)
    end_points.append(x[-1:][0])

plt.subplot(121)
end_points = np.array(end_points)
plt.scatter(end_points[:,0],end_points[:,1],alpha=1,marker='.', c='blue')

N = 100
end_points = list()
for i in range(W):
    _,x = gen_random_walk(N,d)
    end_points.append(x[-1:][0])

plt.subplot(121)
end_points = np.array(end_points)
plt.scatter(end_points[:,0],end_points[:,1],alpha=1,marker='.', c='gray')
```

# Central Limit Theorem

```python
W = 10000
d = 1

np.random.seed(1)
plt.figure(figsize=(12, 3))

a = 1 / np.sqrt(12)

# 1 Step
N = 1
sigma = np.sqrt(N) * a
end_points = np.array([])
for i in range(W):
    _,x = gen_random_walk(N,d)
    end_points = np.append(end_points, x[-1:])

plt.subplot(131)
end_points = np.array(end_points)
_ = plt.hist(end_points, bins=50, range=[-3*sigma, 3*sigma], density=True)

N = 2
sigma = np.sqrt(N) * a
end_points = np.array([])
for i in range(W):
    _,x = gen_random_walk(N,d)
    end_points = np.append(end_points, x[-1:])

plt.subplot(132)
end_points = np.array(end_points)
_ = plt.hist(end_points, bins=50, range=[-3*sigma, 3*sigma], density=True)

N = 100
sigma = np.sqrt(N) * a
end_points = np.array([])
for i in range(W):
    _,x = gen_random_walk(N,d)
    end_points = np.append(end_points, x[-1:])

plt.subplot(133)
end_points = np.array(end_points)
_ = plt.hist(end_points, bins=50, range=[-3*sigma, 3*sigma], density=True)
```
