# Intro
Birthday problem is soveld in 2 methods: 1) Monte Carlo, and 2) using exact analytic formula. The results from 2 methds are compared.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# configs:
N = 100        # monte carlo iterations
M = 50         # max number of the kids
D = 365        # days in a year
```

```python
# Monte Calro simulation:
res = []
for s in range(2,M):
    same_day = 0
    for run_num in range(0, N):
        A = np.random.randint(low=1, high=365, size=s)
        A_sort = np.sort(A)
        ar_diff = A_sort[0:-1] - A_sort[1:]
        if np.max(ar_diff)==0:
            same_day += 1
    res.append(same_day / N)
```

```python
# exact formula:
prob = []
for s in range(2,M):
    tmp_prob = 1
    for i in range(1,s):
        tmp_prob = tmp_prob * (1-i/D)
    prob.append(1-tmp_prob)
```

```python
# Compare the results:
plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(res, c='gray')
plt.plot(prob, c='red')
plt.xlabel('size')
plt.ylabel('probability')
plt.title("Birthday Problem")
plt.legend([f'Monte Carlo N = {N}','Analytic'])
```
